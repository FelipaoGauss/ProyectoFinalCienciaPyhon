##### ¿En que país y que avión se encuentra entre el 85% y el 70% de la cantidad de aterrisajes? Proporcione el nombre del país y el nombre del avión.

```python
# Suponiendo que el origen de los vuelos es igual al país donde se encuentra el aeropuerto
# Primero, necesitamos calcular la cantidad total de vuelos por avión y país
df_flights['total_flights'] = 1  # Agregar columna para contar los vuelos
df_flights['landing_country'] = df_flights['dest']  # Suponemos que el destino del vuelo es el país de aterrizaje

# Agrupar por avión y país y contar la cantidad total de vuelos
df_landings = df_flights.groupby(['tailnum', 'landing_country']).agg(total_flights=('total_flights', 'sum')).reset_index()

# Calcular el total de aterrizajes por avión
df_landings['total_landings'] = df_landings.groupby('tailnum')['total_flights'].transform('sum')

# Calcular el porcentaje de aterrizajes por avión y país
df_landings['landing_percentage'] = df_landings['total_flights'] / df_landings['total_landings'] * 100

# Filtrar los aviones con un porcentaje de aterrizajes entre el 85% y el 70%
filtered_landings = df_landings[(df_landings['landing_percentage'] <= 85) & (df_landings['landing_percentage'] >= 70)]

# Encontrar el país y avión con el porcentaje requerido
result = filtered_landings[['tailnum', 'landing_country']].head(1)

# Imprimir el resultado
print("El avión", result['tailnum'].values[0], "se encuentra en", result['landing_country'].values[0], "con entre el 85% y el 70% de aterrizajes.")
```

##### 2. ¿Indique cual es el areopuerto con la temparatura más alta registrada en los datos?

```python
# Encontrar el aeropuerto con la temperatura más alta
aeropuerto_temperatura_alta = df_weather.loc[df_weather['temp'].idxmax(), 'origin']

# Obtener la temperatura más alta registrada
temperatura_alta = df_weather['temp'].max()

# Mostrar el resultado
print(f"El aeropuerto con la temperatura más alta registrada es {aeropuerto_temperatura_alta}, con una temperatura de {temperatura_alta} grados.")

# Obtener los 3 aeropuertos con las temperaturas más altas
top_airports_temp = df_weather.nlargest(20, 'temp')

# Crear el gráfico de barras
plt.figure(figsize=(12, 8))
plt.bar(top_airports_temp['origin'], top_airports_temp['temp'], color='skyblue')
plt.xlabel('Aeropuerto')
plt.ylabel('Temperatura más alta (°C)')
plt.title('Top 3 Aeropuertos con las Temperaturas más Altas')
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

##### 3. ¿Cuál es la áreaolinea con la menor cantidad de vuelos registrados, indique cuantos vuelos, el código de la áreaolinea, el nombre completo?

```python
# Calcular el número de vuelos por aerolínea
flights_per_airline = df_flights['carrier'].value_counts()

# Obtener la aerolínea con la menor cantidad de vuelos
min_flights_airline = flights_per_airline.idxmin()
min_flights_count = flights_per_airline.min()

# Obtener el nombre completo de la aerolínea
min_flights_name = df_airlines[df_airlines['carrier'] == min_flights_airline]['name'].values[0]

print(f"Aerolínea con la menor cantidad de vuelos:")
print(f"Código de la aerolínea: {min_flights_airline}")
print(f"Nombre completo de la aerolínea: {min_flights_name}")
print(f"Número de vuelos registrados: {min_flights_count}")
```
